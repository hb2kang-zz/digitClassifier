# Question 1

```python
def plotimages(images, Y, scale=0.9, proportion=0.2):
    '''
    Input:
        images: images, must be in a 3-dimensional matrix (x by y by n)
            e.g. if X is 64 by 400 and size of each image is 8 by 8, images=np.reshape(X,(8,8,400));
        Y: coordinates of where to plot the image (Y(1,:) by Y(2,:)
        scale: scale of each image wrt to figure size (scale <= 1, e.g. 0.8)
        proportion: proportion of the data to be ploted (proportion <= 1)
            e.g. if there are 400 data points proportion = 1, plots
            all 400 data points and proportion = 0.5 plot only 200 data points 
            (i.e. 1th, 3th, 5th, ...)
    Ali Ghodsi 2006
    '''
    Y /= norm(Y, axis = 1, ord=2).reshape((Y.shape[0],1))
    inc = np.floor(1/proportion)
    
    image_width = images.shape[0]
    image_height = images.shape[1] 
    n_images = images.shape[2]
    
    plt.gray()
    fig = plt.gcf()
    fig.clf()
    ax = plt.subplot(111)
    ax.set_xlim((Y[0,:].min()*1.2,Y[0,:].max()*1.2))
    ax.set_ylim((Y[1,:].min()*1.2,Y[1,:].max()*1.2))
    for counter in np.arange(0,n_images,inc):
        counter = int(counter)
        xy = (Y[0,counter],Y[1,counter])
        current_image = 1-np.reshape(images[:,:,counter], (image_width,image_height))
        imagebox = OffsetImage(current_image, zoom=scale)
        ab = AnnotationBbox(imagebox, xy, xybox=(1., -1.), xycoords='data', boxcoords="offset points", frameon=False)                                  
        ax.add_artist(ab)
    plt.draw()
    # plt.show()
```

```python
def pca(X, dim):
    u,d,vt = np.linalg.svd(X - np.mean(X, axis=1).reshape((64,1)))
    return np.diag(d[:dim]).dot(vt[:dim,:])
plotimages(images, pca(X, 2))
```

## 1. b) FDA


```python
def fda(X):
    mu = np.mean(X, axis=1)
    mu = mu.reshape(len(mu),1)
    ST = np.average([np.matrix(x-mu).T.dot(x-mu) for x in np.transpose(X)], axis=0)
    
    sigma0 = np.cov(X[:,:100])
    sigma1 = np.cov(X[:,100:200])
    sigma2 = np.cov(X[:,200:])
    SW = np.sum([sigma0, sigma1, sigma2], axis=0)
    
    SB = ST - SW
    
    W = np.linalg.inv(SW).dot(SB)
    w, V = np.linalg.eig(W)
    v0 = np.real(V[:,0])
    v1 = np.real(V[:,1])
    
    x = np.array([x.dot(v0) for x in np.transpose(X)])
    y = np.array([x.dot(v1) for x in np.transpose(X)])
    xy = np.array([y, x])
    return xy

plotimages(images, fda(X))
# x = np.linspace(0, 10, 1000)
# plt.plot(x, 1*np.sin(x*1));
```

let 
$\Sigma_i$ be covariance for class i

$\Sigma_i = \frac{1}{n_i-d} \sum_{x : class\ i} (x-\mu_j)(x-\mu_j)^T$

$\Sigma = \frac{1}{3} \sum_{x = 0}^2 \Sigma_i$

let $\mu_i$ be mean for class i, and in this case, all of they are same

let $\pi_i$ be precondition

let $r_i^2 = (x-\mu_i)\Sigma^{-1}_i(x-\mu_i)$ 

### In LDA:

let $r_i^2 = (x-\mu_i)\Sigma^{-1}(x-\mu_i)$ 

The boundary is $x$ where: 
$r_j^2 + \log |\Sigma| = r_k^2 + \log |\Sigma|$, for $\{i,j,k\} = \{0,1,2\}$


```python
def LDA_boundary(mu_1, mu_2, sigma):
    mu_1 = np.matrix(mu_1.reshape((2,1)))
    mu_2 = np.matrix(mu_2.reshape((2,1)))
    inv = np.linalg.inv(sigma)
    
    const = 0.5* (-mu_1.T.dot(inv).dot(mu_1) + mu_2.T.dot(inv).dot(mu_2))
    v = mu_1.T.dot(inv) - mu_2.T.dot(inv)
    a = -v[0,0]/v[0,1]
    b = -float(const / v[0,1])
    return a, b
    
a,b = LDA_boundary(mu1, mu2, sigma)
print('the boundary for 1 and 2 is y = {}x+{}'
      .format(np.round(a, 2),np.round(b, 2)))

a,b = LDA_boundary(mu0, mu2, sigma)
print('the boundary for 0 and 2 is y = {}x+{}'
      .format(np.round(a, 2),np.round(b, 2)))

a,b = LDA_boundary(mu0, mu1, sigma)
print('the boundary for 0 and 1 is y = {}x+{}'
      .format(np.round(a, 2),np.round(b, 2)))
```

### In QDA:

let $r_i^2 = (x-\mu_i)\Sigma_i^{-1}(x-\mu_i)$

The boundary is $x$ where: $r_j^2 + \log |\Sigma_j| = r_k^2 + \log |\Sigma_k|$, for $\{i,j,k\} = \{0,1,2\}$



```python
def QDA_boundary(mu_1, mu_2, sigma_1, sigma_2):
    mu1 = np.matrix(mu_1.reshape((2,1)))
    mu2 = np.matrix(mu_2.reshape((2,1)))
    inv1 = np.linalg.inv(sigma_1)
    inv2 = np.linalg.inv(sigma_2)
    
    a = inv1 - inv2
    b = mu2.T.dot(inv2) - mu1.T.dot(inv1)
    c = inv2.dot(mu2) - inv1.dot(mu1)
    d = np.log(np.linalg.det(sigma_1)) - np.log(np.linalg.det(sigma_1)) + \
        - mu2.T.dot(inv2).dot(mu2) + mu1.T.dot(inv1).dot(mu1)
    return a, b, c, d

a, b, c, d = QDA_boundary(mu1, mu2, sigma1, sigma2)
print('the boundary for 1 and 2 is x.T * a*x + b*x + x.T*c + d = 0')
print(a)
print(b)
print(c)
print(d)

a, b, c, d = QDA_boundary(mu0, mu2, sigma0, sigma2)
print('the boundary for 1 and 2 is x.T * a*x + b*x + x.T*c + d = 0')
print(a)
print(b)
print(c)
print(d)

a, b, c, d = QDA_boundary(mu0, mu1, sigma0, sigma1)
print('the boundary for 1 and 2 is x.T * a*x + b*x + x.T*c + d = 0')
print(a)
print(b)
print(c)
print(d)
```

```python
data = scipy.io.loadmat('/Users/jianxingzhang/Desktop/stat441a2/faces.mat')
data.keys()
```

```python
X = np.matrix(np.append(data['train_faces'], 
                        data['train_nonfaces'], axis=0)).T
X = np.append(np.matrix(np.ones(np.shape(X)[1])), X, axis=0)
Y = np.append(np.ones(len(data['train_faces'])), 
              np.zeros(len(data['train_nonfaces'])), axis=0)
X_test = np.matrix(np.append(data['test_faces'], 
                             data['test_nonfaces'], axis=0)).T
X_test = np.append(np.matrix(np.ones(np.shape(X_test)[1])), X_test, axis=0)
Y_test = np.append(np.ones(len(data['test_faces'])), 
                   np.zeros(len(data['test_nonfaces'])), axis=0)

beta = np.matrix(np.zeros((np.shape(X)[0], 1)))

def compute_error_rate(X = X, Y = Y):
    correct_count = 0.
    for x, y in zip(X.T, Y):
        if (y == 1 and x.dot(beta) >= 0) or (y == 0 and x.dot(beta) < 0):
                correct_count+=1
    return (len(Y) - correct_count)/len(Y)

error_rate = 0.
while error_rate != compute_error_rate():
    error_rate = compute_error_rate()

    E = [np.exp(float(x.dot(beta))) for x in X.T]
    P = [e/(1+e) for e in E]
    W = np.diag([p*(1-p) for p in P])
    XWX_T = X.dot(W).dot(X.T)
    Xyp = X.dot(Y-P)
    beta += np.linalg.solve(XWX_T, np.matrix(Xyp).T)

print('the training set error rate is: ', compute_error_rate())
print('the testing set error rate is: ', compute_error_rate(X_test, Y_test))
print('first 5 elements of optimum beta are: ')
print(np.squeeze(np.asarray(beta))[:5])
```

## 4. a)

since $\tilde{X} = (I - u_1u_1^T)X$

$\tilde{X}\tilde{X}^T = (I - u_1u_1^T)X((I - u_1u_1^T)X)^T$

$= (I - u_1u_1^T)XX^T(I - u_1u_1^T)^T$

$= (I - u_1u_1^T)XX^T(I - u_1u_1^T)$

$= XX^T - u_1u_1^TXX^T - XX^Tu_1u_1^T + u_1u_1^T XX^T u_1u_1^T$

Since $XX^T$ and $u_1u_1^T$ are symetric matrixs, $u_1u_1^TXX^T$ is symetric and equals $XX^Tu_1u_1^T$

Since $XX^T u_1 = n\lambda_1 u_1$, and $u_1^Tu_1 = 1$

$= XX^T - 2 XX^Tu_1u_1^T + u_1 u_1^T n\lambda_1 u_1 u_1^T$

$= XX^T - 2 n \lambda_1 u_1 u_1^T + n\lambda_1 u_1 u_1^T u_1 u_1^T$

$= XX^T - n \lambda_1 u_1 u_1^T$

Thus,

$\tilde{S} = \frac{1}{n}\tilde{X}\tilde{X}^T  = \frac{1}{n}XX^T - \lambda_1 u_1 u_1^T$
